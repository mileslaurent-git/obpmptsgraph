```python

csvdataio = StringIO(csvdata)
df = pd.read_csv(csvdataio)
xvalue = df['PTS']
yvalue = df['BPM']
player = df['Player']

percentiles = [10 * i for i in range(1, 10)]
px_values = {p: np.percentile(xvalue, p) for p in percentiles}
py_values = {p: np.percentile(yvalue, p) for p in percentiles}
```

```python

def get_color(x, y):
    if x >= px_values[90] and y >= py_values[90]:
        return 'purple'
    elif x >= px_values[80] and y >= py_values[80]:
        return 'blue'
    elif x >= px_values[70] and y >= py_values[70]:
        return 'green'
    elif x >= px_values[60] and y >= py_values[60]:
        return 'yellow'
    elif x >= px_values[50] and y >= py_values[50]:
        return 'orange'
    elif x >= px_values[40] and y >= py_values[40]:
        return 'brown'
    elif x >= px_values[30] and y >= py_values[30]:
        return 'pink'
    elif x >= px_values[20] and y >= py_values[20]:
        return 'grey'
    elif x >= px_values[10] and y >= py_values[10]:
        return 'cyan'
    else:
        return 'white'

colors = [get_color(x, y) for x, y in zip(xvalue, yvalue)]
```

```python

avg_values = {}
for i in range(len(percentiles) - 1):
    p1, p2 = percentiles[i], percentiles[i + 1]
    p_range = df[(xvalue >= px_values[p1]) & (xvalue < px_values[p2]) & (yvalue >= py_values[p1]) & (yvalue < py_values[p2])]
    avg_values[p2] = {
        'PTS': p_range['PTS'].mean(),
        'BPM': p_range['BPM'].mean()
    }
p_90 = df[(xvalue >= px_values[90]) & (yvalue >= py_values[90])]
avg_values[90] = {
    'PTS': p_90['PTS'].mean(),
    'BPM': p_90['BPM'].mean()
}

slope, intercept = np.polyfit(xvalue, yvalue, 1)
regression_line = slope * xvalue + intercept
correlation_coefficient = np.corrcoef(xvalue, yvalue)[0, 1]
residuals = yvalue - regression_line
```

```python

df['Color'] = df.apply(lambda row: get_color(row['PTS'], row['BPM']), axis=1)
df['Residuals'] = residuals

df.to_csv('/Volumes/MilesFiles/Miles Drop.csv', index=False)
```

```python

fig = go.Figure()

# Add scatter plot
fig.add_trace(go.Scatter(
    x=xvalue, y=yvalue,
    mode='markers',
    marker=dict(color=colors, size=10),
    text=[f"Player: {p}<br>PTS: {x}<br>BPM: {y}<br>Residual: {r:.2f}" 
          for p, x, y, r in zip(player, xvalue, yvalue, residuals)],
    hoverinfo='text'
))

# Add regression line
fig.add_trace(go.Scatter(
    x=xvalue, y=regression_line,
    mode='lines',
    line=dict(color='lime'),
    name=f'Regression Line<br>y={slope:.2f}x+{intercept:.2f}'
))



fig.add_vline(x=24, line=dict(color='magenta', dash='dash'), name='x=24')
fig.add_hline(y=0, line=dict(color='white', width=3))

fig.update_layout(
    title='BPM:PTS',
    xaxis_title='PTS (Per 100 Poss)',
    yaxis_title='BPM',
    paper_bgcolor='grey',
    plot_bgcolor='black',
    legend=dict(
        x=1,
        y=1,
        traceorder='normal',
        font=dict(
            family='sans-serif',
            size=12,
            color='white'
        ),
        bgcolor='grey',
        bordercolor='Black',
        borderwidth=2
    )
)

```
